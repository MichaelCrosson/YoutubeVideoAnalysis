```python
# List of search queries (channel names or keywords)
search_queries = ['Mrwhoistheboss', 'Marques Brownlee', 'Arun Maini', 'Austin Evans', 'CNET', 'The Tech Chap',
                  'This is Tech Today', 'Chigz Tech Reviews']
```

```python
# Initialize an empty DataFrame to store combined results
combined_df = pd.DataFrame()
```

```python
# Loop through each search query
for query in search_queries:
    # Perform the search for each query
    searches = yt.search(q=query, max_results=50)
    
    # Convert search results to a temporary DataFrame
    temp_df = pd.DataFrame(searches)
    
    # Initialize lists to store additional metadata for each search result
    view_counts = []
    comment_counts = []
    like_counts = []
    dislike_counts = []
    video_description = []
    
    # Loop through each video_id in the search results to get additional metadata
    for video in searches:
        video_id = video['video_id']
        metadata = yt.get_video_metadata(video_id)
        
        # Append the specific metadata values to the respective lists
        view_counts.append(metadata.get('video_view_count'))
        comment_counts.append(metadata.get('video_comment_count'))
        like_counts.append(metadata.get('video_like_count'))
        dislike_counts.append(metadata.get('video_dislike_count'))
        video_description.append(metadata.get('video_description'))
    
    # Add the new metadata columns to the temporary DataFrame
    temp_df['video_view_count'] = view_counts
    temp_df['video_comment_count'] = comment_counts
    temp_df['video_like_count'] = like_counts
    temp_df['video_dislike_count'] = dislike_counts
    temp_df['video_description'] = video_description
    
    # Append the temporary DataFrame to the combined DataFrame
    combined_df = pd.concat([combined_df, temp_df], ignore_index=True)
```
